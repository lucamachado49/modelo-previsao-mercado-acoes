```python
# 2. Importar bibliotecas
import gradio as gr
from transformers import pipeline
```

```python
# 3. Carregar o pipeline de tradução
translation_pipeline = pipeline("translation", model="Helsinki-NLP/opus-mt-en-roa")
```

```python
# 4. Função de tradução de texto
def translate_text(text):
    result = translation_pipeline(text)
    return result[0]['translation_text']
```

```python
# 5. Função de tradução de áudio (aqui apenas um modelo fictício de transcrição)
def translate_audio_to_text(audio):
    # Simulando a transcrição de áudio para texto
    transcript = "Texto transcrito do áudio"  # Esta parte pode ser substituída por um modelo real de transcrição
    result = translation_pipeline(transcript)
    return result[0]['translation_text']
```

```python
# 6. Função de tradução de documentos
def translate_document(file):
    text = file.read().decode("utf-8")  # Lê o conteúdo do arquivo
    result = translation_pipeline(text)
    return result[0]['translation_text']
```

```python
# 7. Customização da interface com barra de progresso
def translate_with_progress(text):
    with gr.Progress():
        return translate_text(text)
```

```python
# Interface para tradução de texto
text_translation_interface = gr.Interface(
    fn=translate_with_progress,
    inputs=gr.Textbox(lines=2, placeholder='Digite o texto em inglês para traduzir'),
    outputs="text",
    title="Tradutor de Texto",
    description="Traduza textos do inglês para o português",
    theme="default"  # Alterado para "default"
)

# Interface para tradução de áudio
audio_translation_interface = gr.Interface(
    fn=translate_audio_to_text,
    inputs=gr.Audio(type="filepath"),  # Removido o argumento 'source'
    outputs="text",
    title="Tradutor de Áudio",
    description="Grave sua voz em inglês e receba a tradução para o português",
    theme="default"  # Alterado para "default"
)

# Interface para tradução de documentos
document_translation_interface = gr.Interface(
    fn=translate_document,
    inputs=gr.File(label="Carregar documento (txt)"),
    outputs="text",
    title="Tradutor de Documentos",
    description="Faça upload de um documento em inglês (arquivo de texto) para traduzir",
    theme="default"  # Alterado para "default"
)
```

```python
# Criar uma interface com múltiplos componentes (sem usar tabs)
def combined_interface(text, audio, file):
    text_translation = translate_text(text) if text else None
    audio_translation = translate_audio_to_text(audio) if audio else None
    document_translation = translate_document(file) if file else None
    return text_translation, audio_translation, document_translation

# Interface combinada
interface = gr.Interface(
    fn=combined_interface,
    inputs=[
        gr.Textbox(lines=2, placeholder='Digite o texto em inglês para traduzir'),
        gr.Audio(type="filepath"),
        gr.File(label="Carregar documento (txt)")
    ],
    outputs=[
        "text",  # Saída da tradução de texto
        "text",  # Saída da tradução de áudio
        "text"   # Saída da tradução de documentos
    ],
    title="Tradutor Multilíngue",
    description="Traduza textos, áudios e documentos do inglês para o português",
    theme="default"
)
```

```python
# Executar a interface
interface.launch()
```
